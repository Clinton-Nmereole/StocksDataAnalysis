```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib
```

```python
stocks_df = pd.read_csv("indexData.csv")
stocks_df
```

```python
stocks_df.describe()
```

We can create a new column year, by initiallly converting the datatype of stocks_df.Date from object to datetime using the inbuilt pandas function to_datetime. The DatetimeIndex function is then subsequently used to filter just the year from the Date and used to create a new column which we call year.

```python
stocks_df.Date
```

```python
stocks_df['Date'] = pd.to_datetime(stocks_df['Date'])
stocks_df.Date
```

```python
stocks_df["year"] = pd.DatetimeIndex(stocks_df.Date).year
stocks_df.year
```

```python
stocks_df.sort_values(by='High', ascending=False).head(15)
```

```python
plt.bar(stocks_df.year, stocks_df.High)
```

We can use a scatter plot to check the Index performance over the years. To do this, we can simply use the scatterplot function from the seaborn library. The x-axis is the year, the y-axis is the High and for the hue we use the Indexes.

```python
sns.scatterplot(x=stocks_df.year, y=stocks_df.High, hue=stocks_df.Index)
```

```python
stocks_df['Daychange'] = stocks_df.Close - stocks_df.Open 
stocks_df
```

```python
stocks_df['Percent_change'] = (stocks_df.Daychange/ stocks_df.Open) * 100
stocks_df
```

```python
stocks_df.sort_values(by='Percent_change', ascending=False).head(20)
```

```python
stocks_df["month"] = pd.DatetimeIndex(stocks_df.Date).month
stocks_df.month
```

```python
plt.bar(stocks_df.month, stocks_df.High)
```

```python
sns.barplot(x=stocks_df.month, y=stocks_df.Percent_change)
```

From the above graph, we can see that the months with the highest percent increase happened in November and December. This can possibly be attributed to the holidays Thanksgiving in November and Christmas in December. We can also see that the greatest drops happened in the Fall, between August and September.
